```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd

Place = []
Performance = []
Time = []
Price = []

for i in range(1,11):
    url = "https://www.showstart.com/event/list?pageNo=10&pageSize=20&cityCode=10&sortType=9"+str(i)
    quote_page = requests.get(url) 
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for item in soup.find_all("a",attrs={"class":"show-item item"}):            
        place=item.find("div",attrs={"class":"addr"})
        performance=item.find("div",attrs={"class":"title"})
        time=item.find("div",attrs={"class":"time"})
        price=item.find("div",attrs={"class":"price"})
        Place.append(place.text)
        Performance.append(performance.text)
        Time.append(time.text)
        Price.append(price.text)
                
df=pd.DataFrame({"Place":Place,"Performance":Performance,"Time":Time,"Price":Price})           
df           
```

```python
df.to_csv('BEIJINNG_ChaoyangDistrict_LIVEHOUSE.csv', sep='\t', encoding='utf-8')
```

```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd

KTV = []
Info = []

for i in range(1,8):
    url = "https://www.jingcity.com/qiye/ktv/"+str(i)
    quote_page = requests.get(url) 
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for item in soup.find_all("div",attrs={"class":"media-body"}):            
        ktv=item.find("h3",attrs={"class":"mt-0 text-nowrap"})
        info=item.find("span",attrs={"class":"info"})
        KTV.append(ktv.text)
        Info.append(info.text)
                
df=pd.DataFrame({"KTV":KTV,"Info":Info})           
df      
```
